```julia
using PGFPlots
```

```julia
f(x) =  -(x-1)^2 + 3
fp(x) = 2 - 2*x
```

```julia
g(α) = x -> fp(α)*(x - α) + f(α)
```

```julia
x = -1:.05:2
x_mid = (x[1] + x[end])/2
β = -0.25
```

```julia
p = Axis([
    Plots.Linear([x_mid], [f(x_mid)], style="red, mark options={fill=red}",legendentry="\$g_i(x_{mid}, x_{mid})\$"),
    Plots.Linear([x_mid], [g(β)(x_mid)], style="red, mark options={fill=red}", legendentry="\$g_i(x_{mid}, beta) \$"),
    Plots.Linear(x, f.(x), style="blue, mark=none", legendentry="\$f(x)\$"),
    Plots.Linear(x, g(x_mid).(x), style="solid, red, mark=none"),
    Plots.Linear(x, g(β).(x), style="dashed, red, mark=none"),    
    

], legendPos="south east", xlabel="\$x\$")
display(p)
```

```julia
save("midpoint_optimality_pic.tex", p)
```
